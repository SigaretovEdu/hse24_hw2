## Выравнивание на хромосому

Следует выбрать одну хромосому, потому что ресурсы google colab ограничены. В случае проблем, таких как Segmentation fault, error 139, следует выбрать хромосому меньшего размера.

Последовательность нуклеотидов хромосомы можно скачать по адресу https://hgdownload.soe.ucsc.edu/goldenPath/hg38/chromosomes/

```python
!wget https://hgdownload.soe.ucsc.edu/goldenPath/hg38/chromosomes/chr17.fa.gz
!gzip -d chr17.fa.gz
```

```python
!bowtie2-build chr17.fa chromosome_index
```

```python
!mkdir bowtie2_res
```

```python
!echo "$F1 $F2 $FC"
```

```python
%%time
!bowtie2 -p 2 -x chromosome_index -U "$F1.fastq" -S "bowtie2_res/res_$F1.sam"
```

```python
%%time
!bowtie2 -p 2 -x chromosome_index -U "$F2.fastq" -S "bowtie2_res/res_$F2.sam"
```

```python
%%time
!bowtie2 -p 2 -x chromosome_index -U "$FC.fastq" -S "bowtie2_res/res_$FC.sam"
```

```python
!samtools view -h -S -b -o "bowtie2_res/res_$F1.bam" "bowtie2_res/res_$F1.sam"
!samtools view -h -S -b -o "bowtie2_res/res_$F2.bam" "bowtie2_res/res_$F2.sam"
!samtools view -h -S -b -o "bowtie2_res/res_$FC.bam" "bowtie2_res/res_$FC.sam"
```

На этом моменте можно скачать себе на компьютер все полученные .bam или .sam файлы на случай сбоя.

```python
!mkdir macs2
```

```python
!macs2 callpeak --broad -t "bowtie2_res/res_$F1.sam" \
    -c "bowtie2_res/res_$FC.sam" \
 	  -f SAM \
	  --outdir macs2 \
		-n "$F1"
```

```python
!macs2 callpeak --broad -t "bowtie2_res/res_$F2.sam" \
    -c "bowtie2_res/res_$FC.sam" \
 	  -f SAM \
	  --outdir macs2 \
    -n "$F2"
```
